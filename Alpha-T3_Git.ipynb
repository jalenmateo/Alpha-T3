```julia
using Plots
using LinearAlgebra

#functions
Hopping(kx::Float64, ky::Float64, t=1, a=1) = t*(cos(kx*a) + 2*cos(kx*a/2)*cos(ky*a*sqrt(3)/2) + im*(sin(kx*a) - 2*(sin(kx*a/2)*cos(ky*a*sqrt(3)/2))))
HoppingHC(kx::Float64, ky::Float64, t=1, a=1)= t*(cos(kx*a) + 2*cos(kx*a/2)*cos(ky*a*sqrt(3)/2) - im*(sin(kx*a) - 2*(sin(kx*a/2)*cos(ky*a*sqrt(3)/2))))

FAJE=range(-3,3,100)
kx=ky=round.(FAJE;sigdigits=2)
t = 1
a = 1
eig1 = []
eig2 = []
eig3 = []
for i in 1:length(kx)
    H_kernel = [[0 Hopping(kx[i],ky[i]) 0]
        [HoppingHC(kx[i],ky[i]) 0 Hopping(kx[i],ky[i])]
        [0 HoppingHC(kx[i],ky[i]) 0]]
    F = eigen(H_kernel)
    EIGVALS = F.values
#     print(EIGVALS[1])
    push!(eig1,EIGVALS[1])
    push!(eig2,EIGVALS[2])
    push!(eig3,EIGVALS[3])
end

#FUNCTIONS TO GET EIGVAL
function Eigen1(kx::Float64, ky::Float64, t=1, a=1)
    H_kernel = [[0 Hopping(kx,ky) 0]
        [HoppingHC(kx,ky) 0 Hopping(kx,ky)]
        [0 HoppingHC(kx,ky) 0]]
    F = eigen(H_kernel)
    EIGVALS = F.values
    return EIGVALS[1]
end
function Eigen2(kx::Float64, ky::Float64, t=1, a=1)
    H_kernel = [[0 Hopping(kx,ky) 0]
        [HoppingHC(kx,ky) 0 Hopping(kx,ky)]
        [0 HoppingHC(kx,ky) 0]]
    F = eigen(H_kernel)
    EIGVALS = F.values
    return EIGVALS[2]
end
function Eigen3(kx::Float64, ky::Float64, t=1, a=1)
    H_kernel = [[0 Hopping(kx,ky) 0]
        [HoppingHC(kx,ky) 0 Hopping(kx,ky)]
        [0 HoppingHC(kx,ky) 0]]
    F = eigen(H_kernel)
    EIGVALS = F.values
    return EIGVALS[3]
end

surface(kx,ky,Eigen1, xlabel = "kx", ylabel = "ky", zlabel = "E")
surface!(kx,ky,Eigen2)
surface!(kx,ky,Eigen3)
```

```julia
#Quantum spin Hall phase transition
alpha = 1
theta = atan(alpha)
COS = cos(theta)
SIN = sin(theta)
L = 1
lmbda = L/COS

M = [1,0,-1]
mu = 1 #-1
sig = 1 #-1
# function getEigen(alpha, theta, COS, SIN, L, lmbda,m,mu,sig, q)
#     P = lmbda^2 * sin(2*theta)/2 - q^2 - lmbda^2
#     Q = (lmbda^2 + q^2)*lmbda*mu*sig*sin(2*theta)*(COS-SIN)/2
#     E = 2*sqrt(-P/3) * cos(acos((-3*sqrt(3)*Q)/(2*sqrt(-p^3)))/3 - (2*pi*(m-1))/3)
#     return E
# end

P(lmbda, theta, q) = lmbda^2 * sin(2*theta)/2 - q^2 - lmbda^2
Q(lmbda, mu, q, sig, theta, COS, SIN) = (lmbda^2 + q^2)*lmbda*mu*sig*sin(2*theta)*(COS-SIN)/2
E(alpha, theta, COS, SIN, L, lmbda,m,mu,sig, q) = 2*sqrt(-P(lmbda, theta, q)/3) * cos(acos(-3*sqrt(3)*Q(lmbda, mu, q, sig, theta, COS, SIN)/(2*sqrt(-P(lmbda, theta, q)^3)))/3 - (2*pi*(m-1))/3)

q = range(-1,1,100)
```

```julia
plot()
for ii in 1:length(M)
    plot!(q,E.(alpha, theta, COS, SIN, L, lmbda,M[ii],mu,sig, q),legend = true)

end
plot!(title="Square Lattice Band diagrams for Varying t",
xlabel="Momentum",ylabel="Energy")
# display(R.(ks,va[3],wa[3]))
# display(va)
```
